```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('/home/eric/projs/verl/amd_scripts/data/dp.csv')
```

```python
df.head(5)
```

```python
plt.style.use('seaborn-v0_8-paper')
```

```python
plt.plot(df.index, df['dp=1'], marker='', linestyle='-', label='DP=1', color='orange')
plt.plot(df.index, df['dp=2/max'], marker='', linestyle='-', label='DP=2 Maximum', color='darkblue')
plt.plot(df.index, df['dp=2/min'], marker='', linestyle='--', label='DP=2 Minimum', color='darkblue')
plt.plot(df.index, df['dp=4/max'], marker='', linestyle='-', label='DP=4 Maximum', color='darkred')
plt.plot(df.index, df['dp=4/min'], marker='', linestyle='--', label='DP=4 Minimum', color='darkred')


# Add titles and labels for clarity
plt.title('Maximum and Minimum Values Over Time', fontsize=14)
plt.xlabel('RL Step', fontsize=12)
plt.ylabel('Generation Time (Sec)', fontsize=12)

# Add a legend to distinguish the lines
plt.legend(fontsize=10)

# Add grid lines for better readability
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Ensure tight layout
plt.tight_layout()


# Save the plot to a file
plt.savefig('line_chart.png', dpi=300)
```
